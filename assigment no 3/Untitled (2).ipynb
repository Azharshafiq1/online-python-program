```python
"""7.	Write a program which takes 5 inputs from user for different subjectâ€™s marks, total it and generate mark sheet using grades ?"""
eng = int(input("enter the english marks"));
urdu = int(input("enter the urdu marks"));
math = int(input("enter the maths marks"));
isl = int(input("enter the islamiat marks"));
comp = int(input("enter the computer marks"));
total = eng + urdu + math + isl + comp;
print(total);
percentage = total / 500 * 100;
print(percentage);
if percentage < 100 and percentage >= 80:
    print("A GARDE");
elif percentage < 80 and percentage >= 60:
    print("B GRADE");
elif percentage < 60 and percentage >= 40:
    print("C GARDE");
else:
    print("FAIL");
```
